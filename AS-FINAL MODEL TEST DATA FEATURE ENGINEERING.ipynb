```python
import pandas as pd
import numpy as np

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('C:/Users/user/Desktop/pythonfiles/test.csv')
```

```python
dataset.head()
```

```python
cat_features=[features for features in dataset.columns if dataset[features].dtypes=='O' and dataset[features].isnull().sum()>0]
print(cat_features,'', (dataset[cat_features].isnull().sum()/len(dataset))*100)
```

```python
def miss_val(dataset,features):
    data=dataset.copy()
    data[features]=data[features].fillna('missing')
    return data
dataset=miss_val(dataset,cat_features)
```

```python
dataset[cat_features].isnull().sum()
```

```python
num_features=[features for features in dataset.columns if dataset[features].dtypes!='O' and dataset[features].isnull().sum()>0]
print(num_features,'', (dataset[num_features].isnull().sum()/len(dataset))*100)
```

```python
for feature in num_features:
    ## We will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
    
    
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[num_features].isnull().sum()
```

```python
dataset.isnull().sum().any()
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
 
continuous_features=['LotFrontage',
 'LotArea',
 'MasVnrArea',
 'BsmtFinSF1',
 'BsmtFinSF2',
 'BsmtUnfSF',
 'TotalBsmtSF',
 '1stFlrSF',
 '2ndFlrSF',
 'GrLivArea',
 'GarageArea',
 'WoodDeckSF',
 'OpenPorchSF',
 'EnclosedPorch',
 'ScreenPorch']
 
```

```python
for features in continuous_features:
    if 0 in dataset[features].unique():
        pass
    else:
        dataset[features]=np.log(dataset[features])
```

```python
dataset.head()
```

## HANDLING RARE CATEGORICAL FEATURES

```python
categorical_features=[features for features in dataset.columns if dataset[features].dtypes=='O']
categorical_features
```

```python
for features in categorical_features:
    
    temp=dataset[features].value_counts()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[features]=np.where(dataset[features].isin(temp_df),dataset[features],'Rare_val')
    
```

```python
dataset.head()
```

```python
for features in categorical_features:
    dataset[features]=dataset[features].astype('category')
    
```
