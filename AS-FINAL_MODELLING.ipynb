```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
final_train=pd.read_csv('C:/Users/user/Desktop/pythonfiles/FINAL_TRAIN.csv')
final_train.head()
```

```python
x=final_train.drop(['SalePrice'],axis=1)
y=final_train['SalePrice']
```

```python
x.head()
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.50)
```

```python
model=XGBRegressor()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
import math

mse=mean_squared_error(y_test,y_pred)
rmse=math.sqrt(mse)
```

```python
rmse
```
