```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('C:/Users/user/Desktop/pythonfiles/train.csv')
```

```python
dataset.head()
```

```python
cat_missing_var= [features for features in dataset.columns if dataset[features].dtype=='O' and dataset[features].isnull().sum()>0]

print(cat_missing_var,'',(dataset[cat_missing_var].isnull().sum())/len(dataset[cat_missing_var]))
```

```python
def cat(dataset,features):
    data=dataset.copy()
    data[features]=data[features].fillna('Missing')
    return data

dataset=cat(dataset,cat_missing_var)
```

```python
dataset[cat_missing_var]  # all missing objects replaced with missing
```

```python
num_missing_var= [features for features in dataset.columns if dataset[features].dtype!='O' and dataset[features].isnull().sum()>0]
print(num_missing_var,'',(dataset[num_missing_var].isnull().sum())/len(dataset[num_missing_var])*100)
```

```python
for feature in num_missing_var:
    ## We will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
    
    
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[num_missing_var].isnull().sum()
```

```python
dataset.isnull().sum().any()
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
# NOW ALMOST ALL OUR NUMERICAL CONTINUOUS FEATURES HAD SKEWNESS
# OUR AIM IS TO CONVERT THOSE INTO GAUSSIAN FORM USING LOG TRANSFORMATION 
# FOR THIS WE NEED TO SELECT THOSE NUMERICAL VARIABLES WHICH ARE NOT HAVING 0
 
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice'] # non-zero features found in EDA

for features in num_features:
    dataset[features]=np.log(dataset[features])
```

## HANDLING RARE CATEGORICAL FEATURES

```python
# EXTRACTING ALL THE CATREGORICAL FEATURES

cat_numerical=[features for features in dataset.columns if dataset[features].dtype=='O']
cat_numerical
```

```python
# GROUPING EACH CATEGORY AND FINDING WHICH GROUP WITHIN THE CATEGORY HAS COUNT <1% AMD MARKING IT AS RARE VARIABLE ( AS THOSE GROUPS MIGHT NOT BE OF BIG SIGNIFICANCE IN THE DATA)

for features in cat_numerical:
    temp= dataset.groupby(features)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[features]=np.where(dataset[features].isin(temp_df),dataset[features],'rareval')
```

```python
for feature in cat_numerical:
    dataset[feature]=dataset[feature].astype('category')
    
```
